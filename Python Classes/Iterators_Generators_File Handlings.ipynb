yeild() - remembers what is the last thing generated, and what is the logic to perform and rememberes where to stop(end value).

So it will not remember all the values like print() 
we get the same result with print but print will give Nonetype so we will use yeild
we can use return and append to get the datatype in the form of list but without a list we can get with yeild.

```python
# this function is kind of a range function which can be used later.
def gensqr(n):
    for i in range(n):
        yield i**2
gensqr(100000)
```

```python
# here we getting data with the help of yeild so it will gives one by one data
for i in gensqr(10):
    print(i,type(i))
```

```python
def fibi(n):
    a=0
    b = 1
    for i in range(n):
        c = a + b
        a=b
        b=c
        yield c
```

```python
fibi(20)
```

```python
for i in fibi(20):
    print(i)
```

```python
def fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
for i in fib(20):
    print(i,type(i))
```
