```python
#connecting to sql using python.
#creating SQL database and tables in python.

#importing connection module to connect mysql in python.
import mysql.connector   # in local system use pip install mysql-connector-python
#create user 'user'@'%' identified by 'password'

#mydb is a connection variable name it can be any name.
#mydb is a connection between sql and python.
mydb = mysql.connector.connect(
  host="localhost",    #local host like 127.0.0.1
  user="abc",
  password="password"
)
print(mydb.is_connected())   #checks if connection is succeded. code=0(connected), code=1(not connected)

# we need a cursor(pointer) to point each file in python so we are using cursor function.
cur = mydb.cursor()  #pointer

# To reflect the changes
mydb.commit()

# Till here the code is same to make a connection.
```

```python
#we can write any sql command inside the execute function and to see the changes use connectionvariable.commit()

#cursor.exceute('SQL Query') # we can write any sql query iside this function.
cur.execute('Show databases')   #shows number of databases but need to use for loop.

#In python the data will be stored in tuples so we need to use for loop to access the data inside a tuple. ((),(),())
for i in cur:
    print(i)   # To check the rows and columns in python use for loop

cur.execute('Create Database Pythondb1')
#comment create database command once the database is created
cur.execute('use Pythondb1')  
cur.execute('create table pt1(age int, name varchar(40), sex varchar(10))')
#comment create table command once the table is created
cur.execute('insert into pt1 values(23, "krishna lohith", "male") ')
#unless and untill we commit the changes the data won't show up.
cur.execute('select * from pt1')   #To see the results we need to commit and then use for loop.
mydb.commit()  # we need to run this command while not commenting the insert statement - cur.execute('insert into pt1 values(23, "krishna lohith", "male") ')
for i in cur:
    print(i)
#All these are written in .py file. we can check the database and we can also alter it in .sql files
```
